# Problem Statement
Build a sentiment analysis tool that classifies
medical or customer reviews as positive, negative, or neutral
using NLP techniques (TF-IDF + Logistic Regression).

# Dataset Plan
Use the IMDB movie reviews dataset from Kaggle (free).
25,000 labelled reviews — positive and negative.
Will adapt to healthcare review context.

# Approach
Step 1: Load and explore dataset
Step 2: Clean text (lowercase, remove punctuation, remove stopwords)
Step 3: TF-IDF vectorisation
Step 4: Train Logistic Regression classifier
Step 5: Evaluate with precision, recall, F1 score
Step 6: Compare with Naive Bayes baseline
